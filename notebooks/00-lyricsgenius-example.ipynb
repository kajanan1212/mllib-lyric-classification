```python
import pandas as pd
from lyricsgenius import Genius

token = ""

genius = Genius(token, timeout=20)

genre = 'pop'

songs = []

page = 1
while page and len(songs) < 100:
    res = genius.tag(genre, page=page)
    print(res)
    for hit in res['hits']:
        song_lyrics = genius.lyrics(song_url=hit['url'])
        print(song_lyrics)
        print(type(song_lyrics))
        request = genius.search_lyrics(song_lyrics[100:1000])
        print(request)
        print(type(request))
        break
        songs.append(
            {
                "artist_name":  hit['artists'][0],
                "track_name": hit['title'],
                "genre": genre,
                "lyrics": song_lyrics,
            }
        )
    if True: break

    page = res['next_page']

df = pd.DataFrame(songs)
df.to_csv("Student_dataset.csv")
```
